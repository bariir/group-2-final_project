# Spotify Exploratory Data Analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pathlib as Path
import seaborn as sns
```

```python
music_df = pd.read_csv("Resources/spotify_dataset.csv", encoding = "ISO-8859-1")
music_df.head(2)
```

```python
# Find null values in the dataset
pd.isnull(music_df).sum()
```

```python
# Identify total rows and column
music_df.info()
```

```python
music_df['Popularity'].replace('', np.nan, inplace=True)
```

```python
music_df.dropna(subset=['Popularity'], inplace=True)
music_df.head()
```

```python
music_df.info()
```

```python
# Find most popular music
most_popular = music_df.query('Popularity >90', inplace=False).sort_values('Popularity', ascending=False)
most_popular[:10]
```

```python
# Converting duration from millisecond to second on Spotify Music Dataset
music_df['duration'] = music_df['Duration_ms'].apply (lambda x : round(x/1000))
music_df.duration.head()
```

```python
# Make the Release Date column as the index column
music_df.set_index("Release Date", inplace=True)
music_df.index=pd.to_datetime(music_df.index)
music_df.head()
```

```python
# Descriptive statistics of Spotify Music
music_df.describe().transpose()
```

```python
music_df.head(2)
```

```python
music_df.info()
```

```python
# Correlation Heatmap between Variable using Pearson correlation method
ms = music_df.drop(['Index', 'Highest Charting Position','Number of Times Charted','Artist Followers','Duration_ms'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
map = sns.heatmap(ms, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Correlation Heatmap between Variable')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```
